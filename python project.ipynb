```python

import tkinter as tk
from tkinter import *
from tkinter import filedialog
from tkinter.filedialog import askopenfile
import pywhatkit as kt
import pprint
import os
from tkinter import messagebox
import pywhatkit as kt
from PIL import Image, ImageTk


root =Tk()
npth = "C:\\Users\\Dhruvil\\Desktop\\python\\converted_img.txt" 
# naming the GUI interface to image_conversion_APP
root.title("Image_Conversion_App")
 
# creating the Function which converts the jpg_to_png 
def image_to_ascii():
    
    import_filename = fd.askopenfilename()
    im1 = Image.open(import_filename)
    target_path = "converted_img"
    export_filename = kt.image_to_ascii_art(import_filename, target_path)
    im1.save(export_filename)
    

button1 = tk.Button(root, text="Convert", width=20, height=2, bg="green",
                 fg="white", font=("helvetica", 12, "bold"), command=image_to_ascii)
 
button1.place(x=120, y=120)
root.geometry("500x600")


root.mainloop()  # Keep the window open
os.startfile(npth)
```
